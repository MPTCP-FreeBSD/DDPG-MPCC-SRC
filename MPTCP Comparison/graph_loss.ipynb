```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure


df=pd.read_csv("./losspercentagedata/results-trialc.txt", names=["experiment_number","congestion_control","loss_percentage","thrpt_mp","reno_name","thrpt_sp","latency_name","latency"])
```

```python
df
```

```python
plt.rcParams.update({'figure.figsize':(8,6), 'figure.dpi':100, 'font.size':16})
```

```python
train_df=df.copy()
train_df.set_index(df['loss_percentage'])

tcp_group=train_df.groupby(df['congestion_control'])



tcp_list=[]
tcp_list_pd=[]


for group_key, group_value in tcp_group:
    group = tcp_group.get_group(group_key)
    group=group.set_index(group['loss_percentage'])
    group=group.groupby(group.index).mean()
    group=group[['latency']]

    plt.plot(group.index,group['latency'],label=group_key)
    

    # if(group_key=="balia"):
    # group.plot(color="red")

    # fig = plt.gcf()
    # title=group_key
    # plt.title(title)
    # fig.set_size_inches(18.5, 10.5)
    # fig.savefig("Graphs.png", dpi=100,facecolor='white', transparent=False)
            


    tcp_list.append(group_key)
    tcp_list_pd.append(group)


    # group.plot(color="red")
    # from matplotlib.pyplot import figure

    # fig = plt.gcf()
    # title=group_key
    # plt.title(title)
    # fig.set_size_inches(18.5, 10.5)
    # fig.savefig("Graphs.png", dpi=100,facecolor='white', transparent=False)

title="latency vs loss percentage in network a"
plt.title("latency vs loss percentage in network a")
plt.legend()
plt.savefig("./iperf3_tests/"+title+"-"+"Graphs.png", dpi=100,facecolor='white', transparent=False,bbox_inches='tight')
plt.show()
```
