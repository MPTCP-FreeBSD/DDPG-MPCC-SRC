```python
# import pandas as pd

# df=pd.read_csv("results-trialc.txt", names=["experiment_number","congestion_control","bottleneck_buffer","thrpt_t1","thrpt_t2","thrpt_unit","latency"])

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

df=pd.read_csv("./bandwidthdata/results-trialc.txt", names=["experiment_number","congestion_control","bottleneck_buffer","thrpt_mp","reno_name","thrpt_sp","latency_name","latency"])

df['bottleneck_buffer']=df['bottleneck_buffer']/1024

# df['bottleneck_buffer']=pd.Series(np.log10(df['bottleneck_buffer'].to_numpy()))
```

```python
df
```

```python
plt.rcParams.update({'figure.figsize':(8,6), 'figure.dpi':100, 'font.size':16})
```

```python
train_df=df.copy()
train_df.set_index(df['bottleneck_buffer'])

tcp_group=train_df.groupby(df['congestion_control'])





tcp_list=[]
tcp_list_pd=[]


for group_key, group_value in tcp_group:
    group = tcp_group.get_group(group_key)
    group=group.set_index(group['bottleneck_buffer'])
    group=group.groupby(group.index).mean()
    group=group[['latency']]

    plt.plot(group.index,group['latency'],label=group_key)
    

    # if(group_key=="balia"):
    # group.plot(color="red")

    # fig = plt.gcf()
    # title=group_key
    # plt.title(title)
    # fig.set_size_inches(18.5, 10.5)
    # fig.savefig("Graphs.png", dpi=100,facecolor='white', transparent=False)
            


    tcp_list.append(group_key)
    tcp_list_pd.append(group)


    # group.plot(color="red")
    # from matplotlib.pyplot import figure

    # fig = plt.gcf()
    # title=group_key
    # plt.title(title)
    # fig.set_size_inches(18.5, 10.5)
    # fig.savefig("Graphs.png", dpi=100,facecolor='white', transparent=False)

title="latency vs bottleneck in network a"
plt.title("latency vs bottleneck in network a")
plt.legend()
plt.savefig("./iperf3_tests/"+title+"-"+"Graphs.png", dpi=100,bbox_inches='tight')
plt.show()
```
